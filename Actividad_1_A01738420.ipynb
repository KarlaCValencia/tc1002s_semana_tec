# **Actividad de Numpy**

```python
import numpy as np
```

1. Crear un arreglo de 10 cincos

```python
arr_cincos = np.full(10, 5)
print(arr_cincos)
```

2. Crear un arreglo de enteros pares desde 10 hasta 50, inclusive

```python
arr_pares = np.arange(10, 51, 2)
print(arr_pares)
```

3. Crear una matriz cuadrada de 4x4 de números enteros aleatorios

```python
matriz_aleatoria = np.random.randint(1, 10, size=(4, 4))
print(matriz_aleatoria)
```

4. Generar un arreglo de 10 números aleatorios entre 0 y 1 y calcular la suma de todos sus elementos

```python
arr_random_1 = np.random.rand(10)
suma = np.sum(arr_random_1)
print("Arreglo", arr_random_1, "\nSuma:", suma)
```

 5. Generar un arreglo de 10 números aleatorios entre 0 y 1 y calcular su media

```python
arr_random_2 = np.random.rand(10)
media = np.mean(arr_random_2)
print("Arreglo:", arr_random_2, "\nMedia:", media)
```
