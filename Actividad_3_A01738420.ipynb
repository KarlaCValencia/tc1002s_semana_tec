```python
# Se proporciona el nombre del archivo y seleccionamos una hoja del archivo
df = pd.read_excel("/content/Sales Data Presentation - Dashboards.xlsx",
                    sheet_name="Raw Data")
print(df.head())
```

#Limpieza de datos

```python
# Método para detectar NaN en un dataframe
df.isnull().sum()
```

#Agrupamiento y graficación de la actividad

Ejercicio 1

```python
# Agrupar por nombre del producto y sumar la cantidad vendida
productos_menos_vendidos = df.groupby('ProdName')['Quantity'].sum().nsmallest(5)

# Gráfica de barras
productos_menos_vendidos.plot(kind='bar', color='salmon')
plt.title('5 Productos Menos Vendidos')
plt.xlabel('Nombre del Producto')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Ejercicio 2

```python
# Filtrar los datos para ese producto
producto_filtrado = df[df['ProdName'] == 'Scented Olive Oil Lotion Gift Basket']

# Gráfica de caja
plt.boxplot(producto_filtrado['Quantity'])
plt.title('Boxplot - Scented Olive Oil Lotion Gift Basket')
plt.ylabel('Cantidad')
plt.grid(True)
plt.show()
```

Ejercicio 3

```python
# Filtrar los datos de la categoría Bath Products
bath_products = df[df['ProdCategory'] == 'Bath products']

# Agrupar por fecha y sumar la cantidad
ventas_por_fecha = bath_products.groupby('OrderDate')['Quantity'].sum().sort_index()

# Gráfica de línea
ventas_por_fecha.plot(kind='line', marker='o', color='mediumblue')
plt.title('Cantidad de Productos Vendidos - Bath Products')
plt.xlabel('Fecha de Orden')
plt.ylabel('Cantidad Vendida')
plt.grid(True)
plt.tight_layout()
plt.show()
```
