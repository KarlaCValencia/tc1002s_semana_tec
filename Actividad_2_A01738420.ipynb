# **Actividad de Pandas**

```python
import pandas as pd
```

1. Creación e impresión del dataframe ficticio de ventas de frutas


```python
df = pd.DataFrame({'Product': ['Carrots', 'Broccoli', 'Banana', 'Banana', 'Beans', 'Orange', 'Broccoli', 'Banana'],
                   'Category': ['Vegetable', 'Vegetable', 'Fruit', 'Fruit', 'Vegetable', 'Fruit', 'Vegetable', 'Fruit'],
                   'Quantity': [8, 5, 3, 4, 5, 9, 11, 8],
                   'Amount': [270, 239, 617, 384, 626, 610, 62, 90]})
print(df)
```

2. Seleccionar e imprimir las columnas de 'Product' y 'Quantity'

```python
print(df[['Product', 'Quantity']])
```

3. Seleccionar e imprimir los renglones con índices 4 y 6

```python
print(df.iloc[[4, 6]])
```

4. Usar groupby para agrupar por 'Product' y mostrar la suma de la columna 'Quantity'

```python
print(df.groupby('Product')['Quantity'].sum())
```

5. Usar groupby para agrupar por 'Category' y mostrar la suma de la columna 'Amount'

```python
print(df.groupby('Category')['Amount'].sum())
```
